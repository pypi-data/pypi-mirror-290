```python
from hrepr import hrepr
import json

class Test0:
    def __init__(self, data):
        self.data = data

    @classmethod
    def __hrepr_resources__(cls, H):
        return [
            H.script(
                src="https://cdnjs.cloudflare.com/ajax/libs/require.js/2.3.6/require.min.js"
            ),
            H.script("""
                requirejs.config({
                    base: '/static/base',
                    paths: {plotly: "https://cdn.plot.ly/plotly-latest.min.js?noext"}
                });
            """)
        ]
        # return H.script(type="text/javascript", src="https://cdn.plot.ly/plotly-latest.min.js")

    def __hrepr__(self, H, hrepr):
        pid = f"plot{id(self)}"
        return H.inline(
            H.div(id=pid),
            H.script(f"""
            require(["plotly"], function(plotly) {{
                plotly.newPlot(
                    document.getElementById('{pid}'),
                    [{{x: [1, 2, 3, 4, 5], y: [1, 2, 4, 8, 16]}}]
                );
            }});
            """)
        )


class Test:
    def __init__(self, data):
        self.data = data

    @classmethod
    def __hrepr_resources__(cls, H):
        return H.require(
            name="plotly",
            src="https://cdn.plot.ly/plotly-latest.min.js?noext",
        )

    def __hrepr__(self, H, hrepr):
        data = {
            "x": list(range(len(self.data))),
            "y": list(self.data),
        }
        return H.script(
            f"plotly.newPlot(plotdiv, [{json.dumps(data)}]);",
            require="plotly",
            create_div="plotdiv",
        )

hrepr([Test([1, 7, 12, 5, 10, 8, 31])])
```

```python
from hrepr import hrepr

cycode = """
cydiv.style.height = "300px";
cydiv.style.width = "300px";
var cy = cytoscape({

  container: cydiv,

  elements: [ // list of graph elements to start with
    { // node a
      data: { id: 'a' }
    },
    { // node b
      data: { id: 'b' }
    },
    { // edge ab
      data: { id: 'ab', source: 'a', target: 'b' }
    }
  ],

  style: [ // the stylesheet for the graph
    {
      selector: 'node',
      style: {
        'background-color': '#666',
        'label': 'data(id)'
      }
    },

    {
      selector: 'edge',
      style: {
        'width': 3,
        'line-color': '#ccc',
        'target-arrow-color': '#ccc',
        'target-arrow-shape': 'triangle',
        'curve-style': 'bezier'
      }
    }
  ],

  layout: {
    name: 'grid',
    rows: 1
  }

});
"""

class Graph:
    def __init__(self, data):
        self.data = data

    @classmethod
    def __hrepr_resources__(cls, H):
        return H.require(
            name="cytoscape",
            src="https://cdnjs.cloudflare.com/ajax/libs/cytoscape/3.17.0/cytoscape.min.js?noext",
        )

    def __hrepr__(self, H, hrepr):
        
        data = {
            "x": list(range(len(self.data))),
            "y": list(self.data),
        }
        return H.script(
            cycode,
            require="cytoscape",
            create_div="cydiv",
        )

hrepr(Graph({}))
```
